```python
import pandas as pd
```

```python
billboard = pd.read_csv('billboard.csv')
features = pd.read_csv('features.csv')
df = pd.merge(billboard, features, on='pk', how='left', suffixes=('', '_y'))
df.drop([col for col in df.columns if '_y' in col], axis=1, inplace=True)
df.sample(5)
```
